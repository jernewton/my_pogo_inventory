```python
total_prob = 0
p = 1 / 20
for k in range(1, 50):
    prob_first_success = (1 - p) ** (k - 1) * p
    prob_second_success = 1 - (1 - p) ** (49 - k)
    total_prob += prob_first_success * prob_second_success
    print(k, round(total_prob, 4))
```

```python
import random
p = 1 / 20
trials = []
for _ in range(0,1000):
    flag_1 = 0
    flag_2 = 0
    counter = 0
    while(flag_1 == 0):
        rand_val = random.random()
        if(rand_val<p):
            flag_1 = 1
        counter += 1
    while(flag_2 == 0):
        rand_val = random.random()
        if(rand_val<p):
            flag_2 = 1
        counter += 1
        
    trials.append(counter)

summer = 0
for k in range(max(trials)):
    success = trials.count(k)
    summer += success
    if(k%7==0):
        print(k, round(summer/len(trials)*100,3))
```

```python
import random
import matplotlib.pyplot as plt

p = 1 / 20
trials = []

for _ in range(100):
    flag_1 = 0
    flag_2 = 0
    counter = 0
    
    # First process
    while flag_1 == 0:
        if random.random() < p:
            flag_1 = 1
        counter += 1
    
    # Second process
    while flag_2 == 0:
        if random.random() < p:
            flag_2 = 1
        counter += 1

    trials.append(counter)

# Plot histogram
plt.hist(trials, bins=range(min(trials), max(trials)+2), edgecolor='black', align='left')
plt.title("Total Trials to Complete Both Processes")
plt.xlabel("Total Trials")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
